```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report,confusion_matrix
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('./heart_clean.csv')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True, fmt='.1f',cmap = 'Wistia')
plt.title('Confusion Matrix')
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(50, 30))
fig.suptitle('Sample Attributes Visualization', fontsize=50)
plt.style.use('fivethirtyeight')
sns.boxenplot(data['target'], data['cp'], palette = 'rocket', ax=axes[0,0])
axes[0,0].set_title('Relation between chest pain and target', fontsize = 20, fontweight = 30)


sns.violinplot(data['target'], data['chol'], palette = 'colorblind',ax=axes[0,1])
axes[0,1].set_title('Relation of Cholestrol with Target', fontsize = 20, fontweight = 30)

sns.distplot(data['age'], color = 'cyan', ax=axes[0,2])
axes[0,2].set_title('Distribution of Age', fontsize = 20)

sns.swarmplot(data['target'], data['age'], palette = 'rocket', size = 10, ax=axes[1,0])
axes[1,0].set_title('Relation of Age and target', fontsize = 20, fontweight = 30)

dat = pd.crosstab(data['target'], data['restecg']) 
dat.div(dat.sum(1).astype(float), axis = 0).plot(kind = 'bar', 
                                                 stacked = False, 
                                                 color = plt.cm.rainbow(np.linspace(0, 1, 4)), ax=axes[1,1])
axes[1,1].set_title('Relation of ECG measurement with Target', fontsize = 20, fontweight = 30)

sns.boxenplot(data['target'], data['slope'], palette = 'pastel', ax=axes[1,2])
axes[1,2].set_title('Relation between Peak Exercise and Target', fontsize = 20, fontweight = 30)

sns.boxenplot(data['target'], data['ca'], palette = 'mako', ax=axes[2,0])
axes[2,0].set_title('Relation between no. of major vessels and target', fontsize = 20, fontweight = 30)

# sns.catplot(x="target", col="sex", data=data, kind="count",ax=axes[2,1])
sns.boxenplot(data['target'], data['sex'], palette = 'Set3',ax=axes[2,1])
axes[2,1].set_title('Relation of Sex and target', fontsize = 20, fontweight = 30)


size = data['sex'].value_counts()
colors = ['lightblue', 'lightgreen']
labels = "Male", "Female"
explode = [0, 0.01]

my_circle = plt.Circle((0, 0), 0.7, color = 'white')

axes[2,2].pie(size, colors = colors, labels = labels, shadow = True, explode = explode, autopct = '%.2f%%')
axes[2,2].set_title('Distribution of Gender', fontsize = 20)
```

```python
#cp: The chest pain experienced

sns.boxenplot(data['target'], data['cp'], palette = 'rocket')
plt.title('Relation between chest pain and target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
# cholestrol vs target

plt.rcParams['figure.figsize'] = (12, 9)
sns.violinplot(data['target'], data['chol'], palette = 'colorblind')
plt.title('Relation of Cholestrol with Target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15, 5)
sns.distplot(data['age'], color = 'cyan')
plt.title('Distribution of Age', fontsize = 20)
plt.show()
```

```python
# relation between age and target

plt.rcParams['figure.figsize'] = (15, 9)
sns.swarmplot(data['target'], data['age'], palette = 'rocket', size = 10)
plt.title('Relation of Age and target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
# Resting electrocardiographic measurement vs target
  
plt.rcParams['figure.figsize'] = (12, 9)
dat = pd.crosstab(data['target'], data['restecg']) 
dat.div(dat.sum(1).astype(float), axis = 0).plot(kind = 'bar', 
                                                 stacked = False, 
                                                 color = plt.cm.rainbow(np.linspace(0, 1, 4)))
plt.title('Relation of ECG measurement with Target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
# slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)
# checking the relation between slope and target

plt.rcParams['figure.figsize'] = (15, 9)
sns.boxenplot(data['target'], data['slope'], palette = 'pastel')
plt.title('Relation between Peak Exercise and Target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
#ca: The number of major vessels (0-3)

sns.boxenplot(data['target'], data['ca'], palette = 'mako')
plt.title('Relation between no. of major vessels and target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
# relation between sex and target

sns.boxenplot(data['target'], data['sex'], palette = 'Set3')
plt.title('Relation of Sex and target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
size = data['sex'].value_counts()
colors = ['lightblue', 'lightgreen']
labels = "Male", "Female"
explode = [0, 0.01]

my_circle = plt.Circle((0, 0), 0.7, color = 'white')

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors = colors, labels = labels, shadow = True, explode = explode, autopct = '%.2f%%')
plt.title('Distribution of Gender', fontsize = 20)
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()
plt.show()
```
